# Извлечение данных из веб-страниц¶

# Задача 1
Выгрузить недавние новости с сайта https://nplus1.ru/ в датафрейм pandas и сохранить все в файл MS Excel.

```python
import requests 
from bs4 import BeautifulSoup as bs
```

```python
page = requests.get("https://nplus1.ru/")
```

```python
soup = bs(page.text, "html.parser")
```

```python
soup
```

```python
for item in soup.findAll("a"):
    print(item)
```

```python
links = [item["href"] for item in soup.findAll("a")]
print(links)
```

```python
new_links = []
for link in links:
    if link.startswith("/news"):
        new_links += [link]
        
print(new_links)
        
```

```python
new_links = ["https://nplus1.ru" + item for item in new_links]
new_links
```

```python
page_1 = requests.get(new_links[1])
soup_1 = bs(page_1.text, "lxml")
```

```python
metas = [meta for meta in soup_1.find_all("meta")]
print(metas)
```

```python
for meta in soup_1.find_all("meta", {"name": "author"}):
    print(meta)
```

```python
helper = soup_1.find_all("meta", {"name": "author"})[0]
```

```python
helper["content"]
```

```python
def get_data(url):
    session = requests.get(url)
    soup_2 = bs(session.text, "lxml")
    author = soup_2.find_all("meta", {"name": "author"})[0]["content"]
    date = soup_2.find_all("meta", {"itemprop": "datePublished"})[0]["content"]
    return (author, date)
```

```python
from time import sleep
data = []
for link in new_links[:5]:
    data += [get_data(link)]
    sleep(1)
    print(link)
print(data)
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
df = df.rename(columns = {0: "author", 1: "date"})
df
```

```python
df.to_excel("news.xlsx")
```
